```r
num <- c(11,45,78,456,200,45,49,153,8,9)
num
```

```r
get_function <- function(function.name) {
    switch(function.name,
          'mean' = mean,
          'median'= median,
          'sum' = sum,
          )
}
```

```r
get_function('mean')
```

```r
get_function('mean')(num)
```

```r
get_function('median')(num)
```

```r
get_function('sum')(num)
```
